```python
#Importar as bibliotecas
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go 
```

```python
#escolher uma Ação
ticker = yf.Ticker('ITUB4.SA')


#intervalo
df = ticker.history(period='1y')
df
```

```python
#calculo da MM simples 
df['MM5d'] = df['Close'].rolling(window=5).mean().round(2)
df['MM15d'] = df['Close'].rolling(window=15).mean().round(2)
```

```python
df
```

```python
#grafico

candlestick = go.Candlestick(
                            x=df.index,
                            open=df['Open'],
                            high=df['High'],
                            low=df['Low'],
                            close=df['Close']
)

lines_mm5 = go.Scatter(x=df.index,
                       y=df['MM5d'],
                       mode='lines',
                       line_color='blue',
                       name='MM5d')

lines_mm15 = go.Scatter(x=df.index,
                       y=df['MM5d'],
                       mode='lines',
                       line_color='yellow',
                       name='MM5d')

fig=go.Figure(data=[candlestick])
fig.add_trace(lines_mm5)
fig.add_trace(lines_mm15)

fig.show()
```

```python
#calculo d MME = Media Movel Exponencial
df['MME25d']=df['Close'].ewm(span=25).mean()
```

```python
df
```

```python
#cruzamento de 2 Media Moveis 
df['MME30d'] = df['Close'].ewm(span=30).mean()
df['MME100d'] = df['Close'].ewm(span=100).mean()
```

```python
df
```

```python
#grafico
candlestick = go.Candlestick(
                            x=df.index,
                            open=df['Open'],
                            high=df['High'],
                            low=df['Low'],
                            close=df['Close']
)

lines_mm30 = go.Scatter(x=df.index,
                       y=df['MME30d'],
                       mode='lines',
                       line_color='green',
                       name='MME30d')

lines_mme100 = go.Scatter(x=df.index,
                       y=df['MME100d'],
                       mode='lines',
                       line_color='blue',
                       name='MME100d')

fig=go.Figure(data=[candlestick])
fig.add_trace(lines_mm30)
fig.add_trace(lines_mme100)

fig.show()
```
